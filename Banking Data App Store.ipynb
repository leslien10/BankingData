```python
from app_store_scraper import AppStore
import pandas as pd
import numpy as np
```

```python
app_bank = {'vietcombank-smart-otp':'944568124' ,'vietinbank-ipay':'689963454',
            'smartbanking': '1061867449','agribank-e-mobile-banking': '935944952',
            'techcombank-mobile':'1548623362' ,'mb-bank': '1205807363',
            'Tpbank-mobile':'450464147','vpbank-neo':'1209349510',
            'myvib-2-0':'1626624790','sacombank-mbanking': '885814869'}

for bank in app_bank:
    bank_review = AppStore(country='vn', app_name=bank, app_id=app_bank[bank])
    bank_review.review()
    df = pd.DataFrame(np.array(bank_review.reviews),columns=['review'])
    df = df.join(pd.DataFrame(df.pop('review').tolist()))
    df.to_csv('/Users/leslie/Downloads/Banking Data/'+ bank +'_AppleStore.csv')
```
